```python
#Importing all the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#importing dataset
df = pd.read_csv('data.csv')
df.head()
```

```python
df.isnull().sum()
#checking all the null values
```

```python
#dropping all the columns that contains the null values 
df = df.dropna(axis=1)
```

```python
#get a count ofthe number of Malignant (M) or Benign (B) cells
df.diagnosis.value_counts()
```

```python
#visualizing the same 
sns.countplot(df.diagnosis,label="count")
plt.show()
```

```python
df.dtypes
#checking the data types for encoding purpose
```

```python
#encoding  the categorical data values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.iloc[:,1]=le.fit_transform(df.iloc[:,1].values)
```

```python
#Visualizing the pairplot
sns.pairplot(df.iloc[:,1:6],hue="diagnosis")
plt.show()
```

```python
df.head()
```

```python
#correlatoin of the columns
df.iloc[:,1:12].corr()
```

```python
#splitting the data into independent(X)and dependent values(Y)
```

```python
X=df.iloc[:,2:31].values
Y=df.iloc[:,1].values
```

```python
#splitting the dataset into 70% training and 30% testing 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
#Feature scaling 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)

x_test=sc.fit_transform(x_test)
```

```python
#APPLYING THE MODELS 
#LOGISTICREGRESSION
from sklearn.linear_model import LogisticRegression
ll=LogisticRegression()
ll.fit(x_train,y_train)
y_pred=ll.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```

```python
#SUPPORT VECTOR MACHINE(SMV)

from sklearn.svm import SVC
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
y_pred2=svc.predict(x_test)
print(accuracy_score(y_pred2,y_test))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
y_pred3=classifier.predict(x_test)
print(accuracy_score(y_test,y_pred3))
```

```python
from sklearn.metrics import confusion_matrix,classification_report

print(confusion_matrix(y_test,y_pred3))
print(classification_report(y_test,y_pred3))
```

```python
#result shows KNN algorithm is able to classify the record with average of 94% accuracy.
```

```python
#clustering 
from sklearn.cluster import KMeans
import numpy as np

distortion=[]

for i in range(1,10,1):
    km=KMeans(n_clusters=i)
    y_km=km.fit_predict(X)
    distortion.append(km.inertia_)
    print(km.inertia_)

a=np.arange(1,10,1)
plt.plot(a,distortion)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4,max_iter=100,tol=0.0001) 
y_km=km.fit_predict(X)
plt.scatter(X[y_km==0,0],X[y_km==0,1],c='red',marker='s',label='Cluster1')
plt.scatter(X[y_km==1,0],X[y_km==1,1],c='green',marker='*',label='Cluster2')
plt.scatter(X[y_km==2,2],X[y_km==2,3],c='blue',marker='s',label='Cluster1')
plt.scatter(X[y_km==3,2],X[y_km==3,3],c='yellow',marker='*',label='Cluster2')

plt.legend()
plt.grid()
plt.show()
```
